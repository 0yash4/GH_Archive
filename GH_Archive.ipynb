```python
import pandas as pd

# Define the path to your compressed JSON file
file_path = '/content/sample_data/2015-01-01-15.json.gz'

# Read the compressed JSON file into a pandas DataFrame
# The 'compression' parameter is automatically inferred from the file extension (.gz)
df = pd.read_json(file_path, compression='gzip', lines=True)

# Display the first few rows of the DataFrame
df.head()
```

```python
df_normalized_repo = pd.json_normalize(df['repo'])
df_normalized_repo.head()
```

```python
list_df_normalized = [df_normalized_actor, df_normalized_repo, df_normalized_org]
list_df_normalized_str = ["df_normalized_actor", "df_normalized_repo", "df_normalized_org"]
for df1 in list_df_normalized_str:
  name = str(df1).split("_")
  print(name)
```

```python
def rename_Col(list_df_normalized: list[pd.DataFrame], list_df_normalized_str: list[str]):
  for df_dummy, df_name_str in zip(list_df_normalized, list_df_normalized_str):
    rename_Col(df_dummy, df_name_str)
    new_name = df_name_str + "_"
    old_name_list = df_dummy.columns.tolist()
    new_name_list = [new_name + col for col in df_normalized_actor.columns]
    dict1 = dict(zip(old_name_list, new_name_list))
    for key, value in dict1.items():
      print(f"{key} : {value}")
```

```python
def rename_Col_gemini(list_df_normalized: list[pd.DataFrame], list_df_normalized_str: list[str]):
    """
    Renames columns in a list of DataFrames by prepending a given string name.
    Modifies the DataFrames in place.

    Args:
        list_df_normalized (list[pd.DataFrame]): A list of pandas DataFrames.
        list_df_normalized_str (list[str]): A list of strings, where each string
                                            corresponds to the desired prefix for
                                            the columns of the DataFrame at the same index.
                                            Must be the same length as list_df_normalized.
    """
    # Basic validation: Ensure the lists have the same length
    if len(list_df_normalized) != len(list_df_normalized_str):
        raise ValueError("The list of DataFrames and the list of names must have the same length.")

    # Iterate over both lists simultaneously using zip()
    for df_dummy, df_name_str in zip(list_df_normalized, list_df_normalized_str):
        # REMOVE THE RECURSIVE CALL: rename_Col(df_dummy, df_name_str)

        # 1. Construct the prefix (e.g., "actor_")
        prefix = df_name_str.split("_")[2] + "_"

        # 2. Get the current column names of the *current* DataFrame (df_dummy)
        old_name_list = df_dummy.columns.tolist()

        # 3. Create the new column names by prepending the prefix
        #    CORRECTION: Use df_dummy.columns, not df_normalized_actor.columns
        new_name_list = [prefix + col for col in old_name_list]

        # 4. Create the mapping dictionary for renaming
        rename_mapping = dict(zip(old_name_list, new_name_list))

        df_dummy.rename(columns=rename_mapping, inplace=True)
```

```python
rename_Col_gemini(list_df_normalized, list_df_normalized_str)
```

```python
df.drop(columns=['org', 'actor', 'repo'], inplace=True)
```

```python
# Concatenate the normalized DataFrames
df_normalized = pd.concat([df_normalized_actor, df_normalized_repo, df_normalized_org], axis=1)

# Display the first few rows of the new DataFrame
df_normalized.head()
```

```python
df_normalized_cols = [col for col in df_normalized.columns]
print(len(df_normalized_cols))
df_normalized.shape
```

```python
df.head()
```

```python
df_final = pd.concat([df, df_normalized], axis=1)
df_final.head()
```

```python
df_col = df.columns.tolist()
df_final_col_list = []
for col in df_col:
  if col == df_col[-1]:
    for col_normalized in df_normalized_cols:
      df_final_col_list.append(col_normalized)
  else:
    df_final_col_list.append(col)
df_final_col_list.append(df_col[-1])
```

```python
print(df_final_col_list)
```

```python
df_result = df_final[df_final_col_list]
df_result.head()
```

```python
df_result.to_csv('/content/sample_data/result.csv', index=False)
```
